# Generar el CSV

```python
def generate_df_NF019():
    years = [2019, 2020, 2021, 2022]
    entities_data = dict()

    for year in years:
        df = pd.json_normalize(load_complete_data(year, True)['content'])
        # Filtro a la data que queremos analizar
        df = df.loc[(df["tender.status"]=="complete") 
                    & (df["tender.procurementMethod"]=="open")
                    & (df["awards"].notnull()) 
                    & (df["tender.numberOfTenderers"]>0)
                    ].reset_index(drop=True)
        # Creamos columna con número de items por cada "ocid"
        df["tender.numberOfItems"] = df.explode("tender.items"
                                        ).reset_index(drop=True
                                        ).loc[:, ["ocid", "tender.items"]].groupby("ocid"
                                                                            )["tender.items"].count().reset_index()["tender.items"]
        
        # Agrupar por entidad contratante
        df_grouped = pd.DataFrame(df.loc[:, ["ocid", "awards", "buyer.id", "buyer.name", "tender.tenderers", "tender.numberOfTenderers", "tender.numberOfItems"]].groupby("buyer.id"))
        df_grouped.index = df_grouped.loc[:, 0].to_list()
        df_grouped.drop(columns=0, inplace=True)

        for entity in df_grouped.index:
            ocids = df_grouped.loc[entity,1].loc[:, "ocid"].to_list()
            numberOfTenderers = df_grouped.loc[entity,1].loc[:, "tender.numberOfTenderers"].to_list()
            numberOfItems = df_grouped.loc[entity,1].loc[:, "tender.numberOfItems"].to_list()
            empresas = []
            for list_awards in df_grouped.loc[entity,1].loc[:, "awards"].to_list():
                empresas.append([dictionary["suppliers"][0]["id"] for dictionary in list_awards])
                #dictionary = {
                # ...
                #   suppliers: [
                #       {
                #           id: RUC de la empresa
                #           name: nombre de la empresa
                #       }
                #    ]  # Suele haber solo un diccionario en esta lista de suppliers
                # ...
                # }
                
            if entity not in entities_data:
                entities_data[entity] = dict()
                entities_data[entity]["ocids"] = ocids
                entities_data[entity]["numberOfTenderers"] = numberOfTenderers
                entities_data[entity]["numberOfItems"] = numberOfItems
                entities_data[entity]["empresas"] = empresas
            else:
                entities_data[entity]["ocids"] += ocids
                entities_data[entity]["numberOfTenderers"] += numberOfTenderers
                entities_data[entity]["numberOfItems"] += numberOfItems
                entities_data[entity]["empresas"] += empresas

    list_to_df = []
    for entity, data in entities_data.items():
        bidders_per_items = np.array(data["numberOfTenderers"])/np.array(data["numberOfItems"]) # Licitadores por articulos
        mean_per_entity = mean(bidders_per_items)
        
        for i in range(len(data["numberOfTenderers"])):
            flag = int(mean_per_entity > bidders_per_items[i])  # Flag si número medio de licitadores recibidos para un artículo y entidad contratante
                                                                # es mayor al número de licitadores por aticulo en ese proceso de contratación
            ocid = data["ocids"][i]
            for empresa in data["empresas"][i]: # Solo se registran los ganadores
                list_to_df.append([ocid, empresa, flag])

    df_to_return = pd.DataFrame(list_to_df)
    df_to_return.columns = ["ocid", "RUC", "RedFlag_NF019"]
    df_to_return.head()
    
    return df_to_return
```

```python
name_csv = "nf019.csv"
generate_df_NF019().to_csv(url_proyect + '\\data\\redflags_analysis\\' + name_csv, index=False)
```
