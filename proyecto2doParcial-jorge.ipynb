```python
df_2019 = pd.json_normalize(json_2019['content'])
df_2020 = pd.json_normalize(json_2020['content'])
df_2021 = pd.json_normalize(json_2021['content'])
df_2022 = pd.json_normalize(json_2022['content'])
```

```python
df_2019['year'] = '2019'
df_2020['year'] = '2020'
df_2021['year'] = '2021'
df_2022['year'] = '2022'
```

```python
df = pd.concat([df_2019,df_2020,df_2021,df_2022]).reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
df['tender.items']
```

```python
df_tenders = df.loc[:,['id','date','year','ocid','tender.items','tender.value.amount','tender.value.currency','tender.procurementMethod']].dropna(subset=['tender.items']).reset_index(drop=True)
df_tenders = df_tenders.explode('tender.items').reset_index(drop=True)
```

```python
df_tenders.columns
```

```python
df_tenders.columns = ['id','date','year','ocid','tender.items','tender_amount','tender_currency','procurementMethod']
df_tenders
```

```python
df_tenders_items = pd.json_normalize(df_tenders['tender.items'].reset_index(drop=True))
df_tenders_items.columns = ['id_tender', 'quantity', 'description', 'additionalClassifications', 'unit_id',
       'unit_name', 'unit_value_amount', 'unit_value_currency', 'unit_scheme',
       'classification_id', 'classification_scheme',
       'classification_description', 'relatedLot']
df_tenders_items = df_tenders_items.loc[:,['id_tender', 'quantity', 'description', 'classification_id', 'classification_description','unit_value_amount', 'unit_value_currency']]
df_tenders_items
```

```python
df_tenders = pd.concat([df_tenders.drop(columns=['tender.items']),df_tenders_items],axis=1)
df_tenders.head()
```

```python
mean = df_tenders.groupby(by=['classification_id'])['unit_value_amount'].mean()
std = df_tenders.groupby(by=['classification_id'])['unit_value_amount'].std(ddof=0)
```

```python
mean
```

```python
std
```

```python
mean = mean.reset_index()
mean
```

```python
std = std.reset_index()
std
```

```python
mean.columns = ['classification_id', 'mean']
mean
```

```python
std.columns = ['classification_id', 'std']
std
```

```python
df_tenders = df_tenders.merge(mean,on='classification_id')
df_tenders = df_tenders.merge(std, on='classification_id')
df_tenders 
```

```python
df_tenders['zscore'] = (df_tenders['unit_value_amount']-df_tenders['mean'])/df_tenders['std']
```

```python
df_tenders = df_tenders.loc[df_tenders['zscore'].notna()]
```

```python
df_q1 = df_tenders.groupby(by=['classification_id'])['unit_value_amount'].quantile(0.25)
df_q1 = df_q1.reset_index()
df_q1.columns = ['classification_id', 'q1']
df_q1
```

```python
df_q3 = df_tenders.groupby(by=['classification_id'])['unit_value_amount'].quantile(0.75)
df_q3 = df_q3.reset_index()
df_q3.columns = ['classification_id', 'q3']
df_q3
```

```python
df_tenders = df_tenders.merge(df_q1, on='classification_id')
df_tenders = df_tenders.merge(df_q3, on='classification_id')
df_tenders.head()
```

```python
df_tenders['iqr'] = df_tenders['q3'] - df_tenders['q1']
df_tenders.head()
```

```python
df_tenders['NF017'] = 0
df_tenders.loc[(df_tenders["NF017"]< (df_tenders['q1']-1.5*df_tenders['iqr']))|(df_tenders["NF017"]> (df_tenders['q3']-1.5*df_tenders['iqr'])),'NF017'] = 1
df_tenders.loc[:,['ocid','classification_id','classification_description','NF017']].tail(20)
```
