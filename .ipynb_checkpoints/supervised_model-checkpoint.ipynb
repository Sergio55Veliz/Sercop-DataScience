```python
from __init__ import *

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from pandas.core.common import random_state
```

```python
df_companies = pd.read_csv("./data/data_for_supervised_model/companies_contract_info.csv")
df_companies.columns = ["ruc", "total_contacts", "contracts_won"]

df_cluster2 = pd.read_csv("./data/data_for_supervised_model/cluster2_data.csv")
df_cluster2
```

```python
df = df_cluster2.merge(df_companies, on="ruc", how="inner")
df
```

# Undersample

```python
print("Registros por cluster encontrado")
for centroid in df["cluster"].unique():
    print("\tGrupo", centroid, "-> ", df.loc[df["cluster"]==centroid].shape[0])
```

```python
n_undersample = 111
dfg0 = df.loc[df["cluster"]==0].sample(n=n_undersample, random_state=32)
dfg1 = df.loc[df["cluster"]==1]
#dfg2 = df.loc[df["cluster"]==2].sample(n=n_undersample, random_state=32)
```

```python
#dff = pd.concat([dfg0, dfg1, dfg2])
dff = pd.concat([dfg0, dfg1]).reset_index(drop=True)
dff
```

# DataSet

```python
X = dff.iloc[:, [1,2,3,4,6,7]]
y = dff["cluster"]

X.head()
```

```python
# Normalización
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), 
            columns = X.columns)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

# Modelo de Regresión Logística

```python
clf = LogisticRegression(random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
features = X.columns
coef = clf.coef_[0]  # coeficientes del modelo
for i in range(len(features)):
    print(features[i], (14-len(features[i]))*" "+"->", coef[i])
```

```python
clf.score(X_test, y_test)
```

## Evaluation

```python
y_true = y_test
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

sns.set_style('darkgrid')
fig, axs = plt.subplots(figsize=(5,5))
foo = sns.heatmap(cm, # Pivot Table, verificar orden, eso importa para asignar los labels
                  #yticklabels=dias_semana, # Lista | nombre de los labels eje y
                  ax=axs, 
                  cmap='YlOrBr',           # formato de colores -> otros más: https://seaborn.pydata.org/tutorial/color_palettes.html
                  annot=True,              # mostrar los valores
                  fmt='.2f'                # con 2 decimales
                  )

axs.set_xlabel('Predicted Label')
axs.set_ylabel('True label')
foo = axs.set_title('Confusion Matrix')
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score

print("recall     ->", recall_score(y_true, y_pred, average='macro'))
print("precission ->", precision_score(y_true, y_pred, average='macro'))
print("accuracy   ->", accuracy_score(y_true, y_pred))
```
